```python
def check_and_clean_dataset(df, name="Dataset"):
    """
    Check and clean a dataset by handling missing values, duplicates, and data types.
    
    Parameters:
    -----------
    df : pandas.DataFrame
        The dataset to clean
    name : str
        Name of the dataset for reporting purposes
    
    Returns:
    --------
    pandas.DataFrame
        Cleaned dataset
    
    Report includes:
    - Dataset shape
    - Missing values
    - Duplicate rows
    - Data types
    - Basic statistics
    """
    
    print(f"\n{'='*60}")
    print(f"DATA CHECK AND CLEANING REPORT: {name}")
    print(f"{'='*60}\n")
    
    # Original shape
    print(f"Original shape: {df.shape}")
    print(f"Columns: {list(df.columns)}\n")
    
    # Check for missing values
    print("MISSING VALUES:")
    missing = df.isnull().sum()
    if missing.sum() > 0:
        print(missing[missing > 0])
        print(f"Total missing values: {missing.sum()}")
    else:
        print("No missing values found ✓")
    
    # Display missing value percentage
    missing_percent = (df.isnull().sum() / len(df)) * 100
    if (missing_percent > 0).any():
        print("\nMissing value percentages:")
        print(missing_percent[missing_percent > 0])
    
    print("\n" + "-"*60)
    
    # Check for duplicates
    print("DUPLICATE ROWS:")
    duplicates = df.duplicated().sum()
    print(f"Number of duplicate rows: {duplicates}")
    if duplicates > 0:
        print("Removing duplicate rows...")
        df = df.drop_duplicates()
        print(f"New shape after removing duplicates: {df.shape}")
    
    print("\n" + "-"*60)
    
    # Handle missing values
    print("HANDLING MISSING VALUES:")
    for col in df.columns:
        if df[col].isnull().sum() > 0:
            if df[col].dtype in ['float64', 'int64']:
                df[col] = df[col].fillna(df[col].median())
                print(f"  → {col}: Filled with median")
            else:
                df[col] = df[col].fillna(df[col].mode()[0] if len(df[col].mode()) > 0 else 'Unknown')
                print(f"  → {col}: Filled with mode")
    
    print("\n" + "-"*60)
    
    # Data types
    print("DATA TYPES:")
    print(df.dtypes)
    
    print("\n" + "-"*60)
    
    # Basic statistics
    print("BASIC STATISTICS:")
    print(df.describe())
    
    print("\n" + "-"*60)
    print(f"Final shape: {df.shape}")
    print(f"{'='*60}\n")
    
    return df
```

```python
def fix_topics(x):
    if pd.isna(x):
        return []
    x = x.strip("[]").replace("'", "").replace('"', "")
    return [t.strip() for t in x.split(",") if t.strip()]
```

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    roc_auc_score,
    confusion_matrix
)

def evaluate_model(pipeline, X, y, model_name):
    print(f"\n===== {model_name} =====")

    preds = pipeline.predict(X)
    probs = pipeline.predict_proba(X)

    for i, col in enumerate(y.columns):
        y_true = y[col]
        y_pred = preds[:, i]
        y_prob = probs[i][:, 1]

        print(f"\nTarget: {col}")
        print("Accuracy :", accuracy_score(y_true, y_pred))
        print("Precision:", precision_score(y_true, y_pred))
        print("Recall   :", recall_score(y_true, y_pred))
        print("ROC-AUC  :", roc_auc_score(y_true, y_prob))
        print("Confusion Matrix:\n", confusion_matrix(y_true, y_pred))

```

```python
from sklearn.model_selection import learning_curve
import matplotlib.pyplot as plt
import numpy as np

def plot_learning_curve(estimator, X, y, target_name, cv=5, scoring='accuracy', title=None):
    if title is None:
        title = f"Learning Curve: {target_name}"
    
    train_sizes, train_scores, val_scores = learning_curve(
        estimator, X, y,
        cv=cv,
        scoring=scoring,
        n_jobs=-1,
        train_sizes=np.linspace(0.1, 1.0, 5),
        shuffle=True,
        random_state=42
    )
    
    train_mean = np.mean(train_scores, axis=1)
    train_std = np.std(train_scores, axis=1)
    val_mean = np.mean(val_scores, axis=1)
    val_std = np.std(val_scores, axis=1)
    
    plt.figure(figsize=(8,5))
    plt.plot(train_sizes, train_mean, 'o-', color='blue', label='Training score')
    plt.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, alpha=0.1, color='blue')
    
    plt.plot(train_sizes, val_mean, 'o-', color='green', label='Validation score')
    plt.fill_between(train_sizes, val_mean - val_std, val_mean + val_std, alpha=0.1, color='green')
    
    plt.title(title)
    plt.xlabel("Training samples")
    plt.ylabel(scoring)
    plt.legend(loc="best")
    plt.grid(True)
    plt.show()

# Example usage for Random Forest predicting 07-day adoption
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200, random_state=42)

target = "adopted_within_07_days"
X = X_train  # your features without target
y = df_train[target]

plot_learning_curve(rf, X, y, target_name=target)
```

```python
from sklearn.model_selection import learning_curve
import matplotlib.pyplot as plt
import numpy as np

def plot_learning_curve(model, X, y, target_name):
    train_sizes, train_scores, val_scores = learning_curve(
        model, X, y,
        cv=5,                # 5-fold cross-validation
        train_sizes=np.linspace(0.1, 1.0, 5),
        scoring='accuracy',
        n_jobs=-1
    )
    
    train_mean = np.mean(train_scores, axis=1)
    train_std = np.std(train_scores, axis=1)
    val_mean = np.mean(val_scores, axis=1)
    val_std = np.std(val_scores, axis=1)
    
    plt.figure(figsize=(7,5))
    plt.plot(train_sizes, train_mean, 'o-', color='blue', label='Training score')
    plt.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, alpha=0.1, color='blue')
    plt.plot(train_sizes, val_mean, 'o-', color='green', label='Validation score')
    plt.fill_between(train_sizes, val_mean - val_std, val_mean + val_std, alpha=0.1, color='green')
    plt.title(f"Learning Curve: {target_name}")
    plt.xlabel("Training samples")
    plt.ylabel("Accuracy")
    plt.legend(loc="lower right")
    plt.grid(True)
    plt.show()

# Example for Random Forest and Logistic Regression
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

targets = ["adopted_within_07_days", "adopted_within_90_days", "adopted_within_120_days"]

# Assuming X_train and y_train are ready
for target in targets:
    print(f"=== Random Forest: {target} ===")
    rf = RandomForestClassifier(n_estimators=200, random_state=42, n_jobs=-1)
    plot_learning_curve(rf, X_train, y_train[target], target)
    
    print(f"=== Logistic Regression: {target} ===")
    lr = LogisticRegression(max_iter=2000, solver='lbfgs')
    plot_learning_curve(lr, X_train, y_train[target], target)
```
